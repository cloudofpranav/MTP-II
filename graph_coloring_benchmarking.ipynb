# **DSatur**

```python
def deg_sat_algo(G):
    node = list(G.nodes())
    order = len(node)
    color_list = [order + 1] * order
    deg_seq = sorted(G.degree, key = lambda x: x[1], reverse = True)
    n = 0
    vertex_deg_seq = [x[n] for x in deg_seq]
    color_list[0] = 1
    adj = nx.adjacency_matrix(G,nodelist = vertex_deg_seq)
    Natural = np.arange(1,order + 1)
    colored_vertex = []
    s = []
    x = []
    z = []
    for v in range(1, order):
        p = []
        for i in range(1, order):
            for j in range(order):
                if adj[i, j] == 1:
                    if color_list[j] != order + 1:
                        p.append(color_list[j])
                        set(p)
            for n in p:
                x.append(i)
            p.clear()
        for q in x:
            if q not in colored_vertex:
                z.append(q)
        if len(z) != 0:
            w = max(set(z), key = z.count)
        colored_vertex.append(w)
        x.clear()
        for m in range(order):
            if adj[w, m] == 1:
                s.append(color_list[m])
        diff_list = list(set(Natural) - set(s))
        color_list[w] = min(diff_list)
        s.clear()
        diff_list.clear()
        x.clear()
        z.clear()
    v = v + 1
    result = zip(vertex_deg_seq, color_list)
    print("DSatur - Vertex Color Number:", max(color_list))
    return max(color_list)
    #return print(tuple(result))
```

```python
start_time = time.time()
deg_sat_algo(G)
end_time = time.time()
print("Time taken by DSatur =", end_time - start_time)
```

# **Welsh Powell**

```python
def welsh_powell_algo(G):
    node = list(G.nodes())
    order = len(node)
    color_list = [order + 1] * order
    adj = nx.adjacency_matrix(G)
    Natural = np.arange(1,order + 1)
    colored_vertex = []
    K = []
    z = []
    for c in range(order):
        for i in range (order):
            for j in range (order):
                if adj[i, j] == 1:
                    if i not in colored_vertex:
                        z.append(i)
        w = max(set(z),key = z.count)
        for j in range(order):
            if adj[w, j] == 1:
                K.append(color_list[j])
        diff_list = list(set(Natural)-set(K))
        color_list[w] = min(diff_list)
        K.clear()
        diff_list.clear()
        colored_vertex.append(w)
        for j in range(order):
            if adj[w, j] != 1:
                if j not in colored_vertex:
                    if color_list[j] == order + 1:
                        for i in range(order):
                            if adj[i,j] == 1:
                                K.append(color_list[i])
                        if color_list[w] not in K:
                            color_list[j] = color_list[w]
                            colored_vertex.append(j)
                            w = j
            K.clear()
        j = j + 1
    c = c + 1
    result = zip(node, color_list)
    print("Welsh Powell - Vertex Color number:", max(color_list))
    return max(color_list)
    #return print(tuple(result))
```

```python
start_time = time.time()
welsh_powell_algo(G)
end_time = time.time()
print("Time taken by Welsh Powell:", end_time - start_time)
```

# **Function to Create Graph in NetworkX**

```python
def genGraph(n, density):
    G = nx.Graph()
    # We represent the graph by its adjacency matrix
    graph = []
    for i in range(n):
        vertex = []
        for j in range(n):
            if (random.uniform(0, 1) < density):
                vertex.append(1)
            else:
                vertex.append(0)
        graph.append(vertex)

    # Since this is a simple graph, the adjacency matrix must be symmetric
    for i in range(n):
        for j in range(0, i):
            graph[i][j] = graph[j][i]
    for i in range(n):
        graph[i][i] = 0

    # convert adj to Graph object
    for i in range(n):
        G.add_node(str(i))

    for i in range(n):
        for j in range(n):
            if graph[i][j]:
                G.add_edge(str(i), str(j))
    # nx.draw_networkx(G)
    return G
```

# **DSatur Plot Functions**

```python
dsatur_times = []
node_count = []
chrom_numbers = []

density = 0.6
for v in range(5, 101, 5):
    G = genGraph(v, density)
    dsatur_start = time.time()
    chrom_numbers.append(deg_sat_algo(G))
    dsatur_end = time.time()
    print("Time taken for", v, " nodes:", dsatur_end - dsatur_start, "\n")
    dsatur_times.append(dsatur_end - dsatur_start)
    node_count.append(v)

print(dsatur_times)
print(node_count)
```
